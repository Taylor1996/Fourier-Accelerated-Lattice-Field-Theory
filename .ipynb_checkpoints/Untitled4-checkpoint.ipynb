```python
# debugging HMC
m=1.0
omega=1.0
import numpy as np



def U(q):
    U = (1/2.0)*m*omega**2*np.sum(q**2) # total pot energy of lattice
    return U
```

```python
N_tau = 10
q = np.arange(N_tau)
q_shifted  = np.array([q[(i+1)%N_tau] for i in range(len(q))])
sqr_term = np.sum((q + q_shifted)**2)

print(q_shifted)
q=q+1
U(q)
```

```python
def grad_U(q):
    return m*omega**2*q
```

```python
q= np.zeros(10) + 2
grad_U(q)
```

```python
delta_t = 0.1
lf_steps = 10
q_current = np.zeros(10)
p = np.random.normal(0,1,10)
q = q_current
p_current = p

print(p)

# leapfrog
p = p - (grad_U(q) * (delta_t/2.0))

print(p)
print(q)

for i in range(lf_steps):
    q = q + delta_t*(1/m)*(p)
    print(p)
    print(q)

    if i != lf_steps-1:
        p = p - (grad_U(q)* delta_t)

# final leap frog step for momentum
p = p - (grad_U(q) * (delta_t/2.0))
```
