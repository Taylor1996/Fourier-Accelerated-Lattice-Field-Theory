```python
import re
infile = open("output_hot.txt", "r")
s = infile.read()
tokens = s.split('[')[1:]

# we discard tokens[0]
for i,path in enumerate(tokens):
    new_path = re.sub('\]|\n', '', path)
    tokens[i] = new_path
#x_data = [path for path in tokens[1:]]
pathList = [[0]*len(tokens[0])]*len(tokens)
for i in range(len(tokens)):
    pathList[i] = [float(x_i) for x_i in tokens[i].split()]

pathList
```

```python
import numpy as np
import matplotlib.pyplot as plt

psi_0 = np.histogram(pathList,bins='auto')
psi_0
```

```python
plt.scatter(psi_0)
plt.show()
```
